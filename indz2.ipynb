### Hierarchy Clustering

```python
# create dendrogram
plt.figure(figsize = (16, 10));
dendrogram = sch.dendrogram(sch.linkage(airlines, method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='average')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(airlines)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
airlines1 = pd.concat([airlines,Clusters], axis=1)
airlines1
```

```python
airlines1.sort_values("Clusters")
```

### K-Means

```python
kmeans = KMeans(n_clusters=14,random_state=0)
kmeans.fit(airlines)
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1, 19):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(airlines)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 19), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(7, random_state=0)
clusters_new.fit(airlines)
```

### DBSCAN

```python
airlines.head(4)
```

```python
a=airlines.values
a
```

```python
stscaler = StandardScaler().fit(a)
X1 = stscaler.transform(a)
```

```python
dbscan = DBSCAN(eps=0.82, min_samples=6)
dbscan.fit(X1)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['Cluster'])
```

```python
airlines4 = pd.concat([airlines,cl],axis=1)
airlines4
```

```python
a = airlines4['Cluster'].value_counts()
```

```python
a.plot(kind = 'bar')
plt.show()
```
